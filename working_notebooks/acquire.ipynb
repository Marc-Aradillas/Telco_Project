```python
# imports for acquisition phase
import os
import numpy as np
import pandas as pd

from env import get_connect

# wrapper function to make it easier to connect to a database
def get_db_connect(database):
    return get_connect(database)
    
# defined function to query database and read dataframe using pandas
def new_telco_data():
    query = '''
        SELECT *
        FROM customers
        JOIN internet_service_types USING (internet_service_type_id)
        JOIN contract_types USING (contract_type_id)
        JOIN payment_types USING (payment_type_id)
        ;
        '''
    # Reads in dataframe from Codeup database
    df = pd.read_sql(query, get_connect('telco_churn'))
    print('\nTelco dataframe generated.')
    return df

# defined function used to retrieve/search for existing file
def get_telco_data():
    if os.path.isfile('telco.csv'):
        print('Found file')
        # If csv file exists read in data from csv file.
        df = pd.read_csv('telco.csv', index_col=0)
    else:
        print('Retrieving file...\n')
        # Read fresh data from db into a dataframe
        df = new_telco_data()
        # Cache data
        df.to_csv('telco.csv')
        print('\nFile retrieved.')
    return df
```
